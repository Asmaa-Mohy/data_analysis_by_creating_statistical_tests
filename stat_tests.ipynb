##Imports

```python
from statanalyser import StatAnalyser
import numpy as np 
import pandas as pd 
```

## Read data

```python
def wrangle(filepath,sep,include_nulls):
    # Read CSV file
    df = pd.read_csv(filepath,sep=sep)

    if not include_nulls:
      return df
      
    else:
      # drop null values
      df = df.dropna()
      # drop high-cardinality categorical features
      df = df.drop(columns="ID")
    
    return df
```

```python
df1= wrangle("/content/drive/MyDrive/Stat_tests/stat_tests/bank-marketing-uci/bank.csv",";",False)
df2= wrangle("/content/drive/MyDrive/Stat_tests/stat_tests/archive (1)/Train.csv",',',True)
df3= wrangle("/content/drive/MyDrive/Stat_tests/stat_tests/archive (2)/insurance.csv",',',False)
```
